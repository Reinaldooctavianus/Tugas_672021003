```python
#Soal 2: Permen
# Data
nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]

# Menghitung total permen
total_permen = sum(jumlah_permen)

# Menghitung persentase masing-masing permen
persentase_permen = [(jumlah / total_permen) * 100 for jumlah in jumlah_permen]

# Plotting Pie Chart dengan label persentase
plt.figure(figsize=(8, 6))
labels = [f'{nama} ({persen:.1f}%)' for nama, persen in zip(nama_permen, persentase_permen)]
plt.pie(jumlah_permen, labels=labels, autopct='%1.1f%%')
plt.title('Persentase Jumlah Permen Berdasarkan Jenis')
plt.show()

# Menampilkan persentase khusus untuk Kopiko
indeks_kopiko = nama_permen.index('Kopiko')
print(f'Peluang memilih permen Kopiko : {persentase_permen[indeks_kopiko]:.1f}%')
```

```python
# Soal 3: Makanan
# Data penjualan makanan penutup
datapenjualan_makananpenutup = {
    'Donat': 14, 'Pastel': 5, 'Kue Coklat': 12, 'Ice Cream': 19,
    'Puding Vanila': 6, 'Brownies': 8, 'Puding Strawberi': 12,
    'Puding Coklat': 9, 'Ice Cream Nutela': 10, 'Kue Coklat-Keju': 17,
    'Kue Wajik': 2, 'Kue Sus': 9, 'Mochi': 13
}

# Sorting data berdasarkan jumlah penjualan
sorted_datapenjualan = dict(sorted(datapenjualan_makananpenutup.items(), key=lambda item: item[1], reverse=True))

# Ambil nama makanan penutup dan jumlah penjualannya
nama_makanan = list(sorted_datapenjualan.keys())
jumlah_penjualan = list(sorted_datapenjualan.values())

# Plotting bar chart
plt.figure(figsize=(10, 6))
plt.bar(nama_makanan, jumlah_penjualan)
plt.xlabel('Makanan Penutup')
plt.ylabel('Jumlah Penjualan')
plt.title('Popularitas Relatif Makanan Penutup (Bar Chart)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Rekomendasi 3 makanan penutup terendah yang dihapus
makanan_terendah = nama_makanan[-3:]
print('Makanan penutup yang direkomendasikan untuk dihapus:', makanan_terendah)

plt.show()
```
